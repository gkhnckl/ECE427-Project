## 2. Analysis of changes in the Wisconsin grid

To no surprise, Wisconsin’s (WI) electricity generation was, and still is, predominantly sourced from coal.  In 2001, energy production in WI was 70% coal based as in Figure 1. This share has since declined to 39% in 2020, at an average rate of -1.6% per year.  Coal remains the primary fuel source for generation in WI, but this will change.  

However, historical data is not showing this change to be a positive one [1]. This is because although the state’s use of coal reduced, its reliance on natural gas grew.  Electricity generation from natural gas in WI increased from 4% in 2001 to 36% in 2020, at an average rate of 1.6% per year [1].  The replacement of coal with natural gas has concerning implications regarding WI’s progress towards a carbon-neutrality.  While natural gas may help the state transition to a sustainable future, it should not be part of it [1]. 



In 2019, under Executive Order No. 38, Governor Tony Evers created the Office of Sustainability and Clean Energy (OSCE) [2].  The goal of “ensuring all electricity consumption within the State of Wisconsin is 100 percent carbon-free by 2050” is declared in the Order. This policy and the Clean Energy Plan that is under development, will drastically change WI’s energy infrastructure, one of which involves the expansion of renewable energy generation.


### 2.1 Generation retirements

The reduction of coal use in WI will continue as more coal plants are retired.  In 2005, WI had a coal generation capacity of 7,123 MW [3].  It has decreased by 18% to 5,846.4 MW in 2020 [1].  Table 1 lists the seven existing coal power plants in WI, most of which will retire within the next two decades.  Alliant Energy has announced that its Edgewater and Columbia plants will close by 2024 [4] [5].  In addition, coal generation at Weston will cease by 2023 [6] and at South Oak Creek by 2035 [7].  As a result, 2,135.2 MW of coal generation capacity will remain.  





Figure 2 illustrates the change of coal plants in Wisconsin. Seven existing coal plants in 2020, spanning from the west to east coasts, will be limited to the borders after 2035 with only two remaining.



Utility companies in WI are motivated to retire coal in order to meet their sustainability goals.  Alliant Energy pledged to completely eliminating coal from its generation fleet by 2040 and reaching net-zero production by 2050 [8].  WEC Energy Group, WI’s largest utility company, intends to cut coal by 2035 [9] and also reaching carbon-neutrality by 2050.  Both companies plan to replace coal with cleaner sources, which may include natural gas.

Figure 3 shows the natural gas territories in Wisconsin. WEC Energy Group has investigated converting the 1,100 MW Columbia facility into a natural gas one [9].  The company is committed to “improving [its] natural gas distribution system,” not ending it [9].  This improvement will aim for zero methane emissions by 2030.  On the one hand, natural gas allows for the rapid retirement of coal and can serve as the baseload while carbon-neutral power plants are integrated into the grid. However, natural gas is not a good option for electricity generation due to its highly-explosive properties and high maintenance costs. Therefore, it is imperative to identify other alternatives. 

In our proposal, we first focus on phasing out coal. That means our proposed model does not include any coal generation within the state, totaling a loss of 415 MW. Note that this figure does not refer to the total nameplate capacity of existing coal plants that we are planning to remove, but their actual generation, depending on the current demands, which can be read from the file Wisconsin_Existing.m. 

      


### 2.2 Generation and transmission additions

Considering the sustainable low-carbon energy demand and ongoing effort on switching to renewable energy resources, for the future of WI grid we look to wind and solar as promising solutions, envisioning an 100% carbon-free electricity generation. Given a generation curtailment of 415 MW due to the retirement of all existing coal plants, we plan to introduce two new ones, each sourced from wind and solar energy. Their profiles will be delineated by a combination of factors, including the growing demand, natural conditions, and our pursue of high cost-effective performance.

Sustainable and reliable energy will involve renewables like hydro, solar and wind. These renewables provide low-carbon power generation, and their generation capacity is increasing every year. The increasing trend on renewable energy resources has taken place in Wisconsin as well. In Figure 5, the annual generation of renewable energy resources in Wisconsin is shared by different renewables aforementioned [12]. Among them, both solar and wind generation are increasing. Compared to coal and natural gas generation, the renewable energy provides sustainable solution. ### 3.1. Wind plant addition


Wisconsin has great wind potential due to its geography.  Figure 6 illustrates wind speeds throughout the state at 80 m above ground level.  The largest average wind speeds (6.5 to 7.5 m/s) are mainly concentrated in the southeast side of Wisconsin, along Lake Michigan. Similar speeds can be found in central Wisconsin and at the westernmost border. Therefore, siting of the proposed wind farm targets the southeastern coast of Wisconsin.   



Presently, no wind farm of similar scale exists in this region. Figure 7 is a map of utility wind farms in Wisconsin as of 2019, among which the largest land-based one is called Glacier Hills. Having a capacity of 162 MW, it is owned by WEC Energy Group and located 100 miles northwest of Oak Creek, which occupies about 17,000 acres for all 90 turbines with 1.8 MW each.



In our proposal, retired coal plants at Oak Creek, a city sitting on the southwestern shore of Lake Michigan, will be replaced with a single 200 MW wind farm. With a rated voltage of 18 kV, it is directly connected to the Oak Creek generator bus (bus #20). From the land-use perspective, we predict our proposed wind farm will require less land use by utilizing more powerful turbines. This is because some of the most advanced wind turbines have a rating of 9-10 MW [14], which means our wind plant will only require 20-23 turbines, thus not so much space as Glacier Hills is needed. Given a 200 MW wind plant has been constructed in Cassia County, Idaho [15], we can take it as a reference and assume our 200 MW plant will be built on a land with the same area. As selected in Figure 8 (the map of Oak Creek county), the region looks flat and available. Therefore, we add our wind plant here which is represented in Figure 9.

## 4. Power flow analysis

As discussed in the project description, the proposed grid model should be able to operate in two conditions: a stormy winter day with little sun and a sunny summer day with a lot of sun. For the analysis, we assumed that the total loads are the same for both cases. In the summer case, we assumed that the solar plant is operating at 100% capacity, and wind plant is operating at 70% capacity. Similarly, for the winter case, we assumed that the solar plant is operating at 70% capacity, and the wind plant is operating at 100% capacity. This assumptions for two cases are tabulated in Table 4. 



Since we assumed that the total load is the same for both cases, we should also have the same generation for both cases for a balanced generation/consumption. As seen in Table 4, the total generation is 340 MW for winter and summer cases. 

Before sharing the analysis results, we can remind how “feasible solution” is defined. As explained in the project description, we should satisfy the following requirements to achieve a feasible solution for a defined load and generation.

•	The generators should not be overloaded.

•	The transmission lines should not be overloaded.

•	The bus voltages should be between 0.95 – 1.05 pu.

If the modeled grid satisfies those three conditions for both winter and summer cases, we can achieve our goal. In the following sections, the analysis results for winter and summer cases are shared.

```julia
# Load the Wisconsin data file
proposed_network = PowerModels.parse_file("Wisconsin_Proposed.m");
```

```julia
# Run a power flow
solve_power_flow!(proposed_network)

# Plot the power flow results
plot_powerflow(proposed_network)

# Compute the total load, generation and losses
total_load = 0.0
for i in keys(proposed_network["load"]) 
    total_load=total_load+proposed_network["load"][i]["pd"]*proposed_network["baseMVA"]
end

total_gen = 0.0
for i in keys(proposed_network["gen"]) 
    total_gen=total_gen+proposed_network["gen"][i]["pg"]*proposed_network["baseMVA"]
end

println("Total load = $total_load MW, total generation = $total_gen MW and total losses = $(total_load-total_gen) MW ")
```

#### Scale the active and reactive power of the load by a scale factor
Here, we scaled the active and reactive power of the loads by 2% by considering the estimated consumption in the future of Wisconsin.

```julia
# Check total load active and reactive power BEFORE scaling


# Compute the total load active and reactive power 
total_p_load = 0.0
for i in keys(proposed_network["load"]) 
    total_p_load=total_p_load+proposed_network["load"][i]["pd"]*proposed_network["baseMVA"]
end

total_q_load = 0.0
for i in keys(proposed_network["load"]) 
    total_q_load=total_q_load+proposed_network["load"][i]["qd"]*proposed_network["baseMVA"]
end

println("Before Scaling:: Total load active power = $total_p_load MW, Total load reactive power = $total_q_load MVAR, ")
```

```julia
# Change the Active and Reactive Power at each bus by a scale factor

k_scale = 1.02    # scale ratio for active and reactive power on loads

scale_loads!(proposed_network, k_scale)
```

```julia
# Check total load active and reactive power AFTER scaling


# Compute the total load active and reactive power 
total_p_load = 0.0
for i in keys(proposed_network["load"]) 
    total_p_load=total_p_load+proposed_network["load"][i]["pd"]*proposed_network["baseMVA"]
end

total_q_load = 0.0
for i in keys(proposed_network["load"]) 
    total_q_load=total_q_load+proposed_network["load"][i]["qd"]*proposed_network["baseMVA"]
end

println("After Scaling:: Total load active power = $total_p_load MW, Total load reactive power = $total_q_load MVAR, ")
```
